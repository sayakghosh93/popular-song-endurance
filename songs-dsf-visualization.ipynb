```python
import ast
import numpy as np

def list_genre(row):
    try:
        #print(row['genre'])
        genre = ast.literal_eval(row['genre'])
        return genre[0]
    except:
        return np.nan

df['main_genre'] = df.apply(list_genre, axis=1)
df['main_genre'].map(lambda x: str(x))
```

```python
df['main_genre'].isna().sum()
```

```python
    def simplify_genre(df):
        hip_hop = ['bass trap']
        rock = ['emo','permanent wave','british invasion','mellow gold','jam band']    
        r_and_b = ['bachata','doo-wop','motown','neo mellow','classic soul','funk']
        dance = ['cabaret']
        fusion = ['comic']
        jazz = ['bebop']
        pop = ['latin']
        metal = ['redneck']

        for i,row in df.iterrows():
            print(row['main_genre'])
            if (('rock' in row['main_genre']) | (row['main_genre'] in rock) ):
                df.at[i,'main_genre'] = 'rock'
            elif (('pop' in row['main_genre'])| (row['main_genre'] in pop)):
                df.at[i,'main_genre'] = 'pop'
            elif ('folk' in row['main_genre']):    
                df.at[i,'main_genre'] = 'folk'
            elif (('metal' in row['main_genre'])|(row['main_genre'] in metal)):    
                df.at[i,'main_genre'] = 'metal'
            elif (('hip hop' in row['main_genre'])|(row['main_genre'] in hip_hop)):    
                df.at[i,'main_genre'] = 'hip hop'
            elif (('jazz' in row['main_genre'])|(row['main_genre'] in jazz)):    
                df.at[i,'main_genre'] = 'jazz'
            elif ('funk' in row['main_genre']):    
                df.at[i,'main_genre'] = 'funk'
            elif ('indie' in row['main_genre']):    
                df.at[i,'main_genre'] = 'indie'
            elif ('rap' in row['main_genre']):    
                df.at[i,'main_genre'] = 'rap'
            elif (('dance' in row['main_genre'])|(row['main_genre'] in dance)):    
                df.at[i,'main_genre'] = 'dance'
            elif ('gospel' in row['main_genre']):    
                df.at[i,'main_genre'] = 'gospel'
            elif ('adult standards' in row['main_genre']):    
                df.at[i,'main_genre'] = 'oldies'
            elif ('country' in row['main_genre']):    
                df.at[i,'main_genre'] = 'country'        

```
